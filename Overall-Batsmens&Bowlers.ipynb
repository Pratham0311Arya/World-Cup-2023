```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
#https://www.espncricinfo.com/records/tournament/icc-cricket-world-cup-2023-24-15338
```

# Batsmen Data

```python
url = "https://www.espncricinfo.com/records/tournament/batting-most-runs-career/icc-cricket-world-cup-2023-24-15338"
req = requests.get(url).text
sp_data = BeautifulSoup(req,"lxml")
```

```python
batsmen = []
table = sp_data.find("tbody")
players = table.find_all('tr')


for i in players:
    bat = i.find_all('td')
    
    singlebatsmen = [j.text for j in bat]
    batsmen.append(singlebatsmen)
```

```python
head = []
t_head = sp_data.find("thead",class_="ds-bg-fill-content-alternate ds-text-left")
headings = t_head.find_all("td")

for i in headings:
    head.append(i.text)   
```

```python
df = pd.DataFrame(batsmen)
```

```python
df.columns=head
```

```python
df
```

```python
df['Team'] = df['Player'].apply( lambda x: x.split("(")[1].split(")")[0])
```

```python
df['Player'] = df['Player'].apply( lambda x: x.split("(")[0])
```

```python
df.drop("Span",axis=1,inplace=True)
```

```python
df.rename({"Mat":"Matches","Inns":"Innings","Ave":"Batting Average","HS":"Highest Score","BF":"Balls Faced"},axis=1,inplace=True)
```

```python
df
```

```python
df.columns
```

```python
batters = df[['Player','Team','Matches', 'Innings', 'NO', 'Runs', 'Highest Score','Batting Average', 'Balls Faced', 'SR', '100', '50', '0', '4s', '6s']]
```

```python
batters.to_excel("batters.xlsx")
```

# Bowlers Data

```python
url2 = "https://www.espncricinfo.com/records/tournament/bowling-most-wickets-career/icc-cricket-world-cup-2023-24-15338"
req2 = requests.get(url2).text
sp_data2 = BeautifulSoup(req2,"lxml")
```

```python
bowlers = []
table = sp_data2.find("tbody")
players2 = table.find_all('tr')


for i in players2:
    bowl = i.find_all('td')
    
    singlebowler = [j.text for j in bowl]
    bowlers.append(singlebowler)
```

```python
head2 = []
t_head2 = sp_data2.find("thead",class_="ds-bg-fill-content-alternate ds-text-left")
headings2 = t_head2.find_all("td")

for i in headings2:
    head2.append(i.text)   
```

```python
head2
```

```python
df2 = pd.DataFrame(bowlers)
```

```python
df2.columns=head2
```

```python
df2
```

```python
df2['Team'] = df2['Player'].apply( lambda x: x.split("(")[1].split(")")[0])
```

```python
df2['Player'] = df2['Player'].apply( lambda x: x.split("(")[0])
```

```python
df2.drop("Span",axis=1,inplace=True)
```

```python
df2.rename({"Inns":"Innings","Mdns":"Maidens","Mat":"Matches","Wkts":"Wickets","BBI":"Best","Ave":"Average","Econ":"Economy"},axis=1,inplace=True)
```
