```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.espncricinfo.com/series/icc-cricket-world-cup-2023-24-1367856/match-schedule-fixtures-and-results"
req = requests.get(url).text
```

```python
sp_data = BeautifulSoup(req,'lxml')
```

```python
sp_data.find('div',class_="ds-grow ds-px-4 ds-border-r ds-border-line-default-translucent").find('a').get('href')
```

```python
divs = sp_data.find_all('div',class_="ds-grow ds-px-4 ds-border-r ds-border-line-default-translucent")
```

```python
links = []
for i in divs:
    link = i.find('a').get('href')
    links.append("https://www.espncricinfo.com/"+link)
```

```python
for i in links:
    print(i)
```

```python
batsmen = []
bowler = []

def extract_data(links):
    for i in links:
        req = requests.get(i).text
        souped = BeautifulSoup(req,'lxml')
        
        tables = souped.find_all('table')
        
        for i in tables[0]:
            tr = i.find_all("tr")
    
            for j in tr:
                td = j.find_all("td")
        
                singlebatsmen = [j.text for j in td]
                batsmen.append(singlebatsmen)
        
        for i in tables[2]:
            tr = i.find_all("tr")
    
            for j in tr:
                td = j.find_all("td")
        
                singlebatsmen = [j.text for j in td]
                batsmen.append(singlebatsmen)
        

        for i in tables[1]:
            tr = i.find_all("tr")
    
            for j in tr:
                td = j.find_all("td")
        
                singlebowler = [j.text for j in td]
                bowler.append(singlebowler)
            
        for i in tables[3]:
            tr = i.find_all("tr")
    
            for j in tr:
                td = j.find_all("td")
        
            singlebowler = [j.text for j in td]
            bowler.append(singlebowler)      
```

```python
extract_data(links)
```

```python
print(len(batsmen))
print(len(bowler))
```

```python
batsmens = []
bowlers = []

for i in batsmen:
    if(len(i)==8):
        batsmens.append(i)
    else:
        pass

bowlers = []

for i in bowler:
    if(len(i)==11):
        bowlers.append(i)
    else:
        pass
```

```python
print(len(batsmens))
print(len(bowlers))
```

```python
bat = pd.DataFrame(batsmens)
bowl = pd.DataFrame(bowlers)

bhead = ['Batsmen','Out-by','R','B','M','4s','6s','SR']
bohead = ['Bowler','O','M','R','W','Eco','0s','4s','6s','WD','NB']
```

```python
bat.columns = bhead
bowl.columns = bohead
```
