```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.cricbuzz.com/cricket-series/6732/icc-cricket-world-cup-2023/matches"
req = requests.get(url)
req
```

```python
src_data = req.text
```

```python
souped_data = BeautifulSoup(src_data,'lxml')
```

```python
divs = souped_data.find_all('div',class_="cb-col-60 cb-col cb-srs-mtchs-tm")
```

```python
urls = []
for i in divs:
    link = i.find('a')
    href = "https://www.cricbuzz.com/" + link.get('href')
    urls.append(href)
```

```python
matches_data = []
for i in urls:
    req = requests.get(i).text    
    souped =  BeautifulSoup(req,'lxml')
    
    teams = souped.find('div',class_="cb-nav-main cb-col-100 cb-col cb-bg-white").find('h1').text
    
    scores = souped.find('div',class_="cb-col cb-col-100 cb-col-scores")
    if scores:
        t1 = scores.find('div',class_="cb-col cb-col-100 cb-min-tm cb-text-gray").text
        t2 = scores.find('div',class_="cb-col cb-col-100 cb-min-tm").text
    else:
        t1,t2 = None,None
        
    loc = souped.find('div',class_="cb-nav-subhdr cb-font-12")
    venue = loc.find_all('span')
    location = venue[5].text
    time = venue[8].text
    stadium = venue[4].text
    
    result = souped.find('div',class_="cb-col cb-col-100 cb-min-stts cb-text-complete").text
    
    motm = souped.find("div",class_="cb-col cb-col-67 cb-nws-lft-col cb-comm-pg").find('div',class_="cb-col cb-col-50 cb-mom-itm").find("a").text
    
    match = {
        "Teams":teams,
        "t1":t1,
        "t2":t2,
        "venue":location,
        "time":time,
        "stadium":stadium,
        "result":result,
        "Man of the match":motm
    }
    
    matches_data.append(match)
    
```

```python
len(matches_data)
```

```python
df = pd.DataFrame(matches_data)
```

```python
df
```

```python
df["MNo."] = df['Teams'].apply(lambda x: x.split(",")[1].split("-")[0].strip())
```

```python
df["Team-1"] = df["t1"].apply(lambda x : x.split(" ")[0].strip())
```

```python
df["Team-2"] = df["t2"].apply(lambda x : x.split(" ")[0].strip())
```

```python
df.drop("Teams",axis=1,inplace=True)
```

```python
df["T1-Runs"] = df["t1"].apply(lambda x: x.split(" ")[1].split("/")[0])
```

```python
df["T1-W"] = df["t1"].apply(lambda x: x.split(" ")[1].split("/")[1] if len(x.split(" ")) > 1 and len(x.split("/")) > 1 else "10")
```

```python
df["T2-Runs"] = df["t2"].apply(lambda x: x.split(" ")[1].split("/")[0])
```

```python
df["T2-W"] = df["t2"].apply(lambda x: x.split(" ")[1].split("/")[1] if len(x.split(" ")) > 1 and len(x.split("/")) > 1 else "10")
```

```python
df.drop(["t1","t2"],axis=1,inplace=True)
```

```python
df['stadium'] = df['stadium'].apply(lambda x: x.split(",")[0].strip())
```

```python
df['Date'] = df["time"].apply(lambda x: x.split(",")[0])
```

```python
df['Time'] = df["time"].apply(lambda x: "Afternoon"if "02:00" in x else "Morning")
```

```python
df.drop("time",axis=1,inplace=True)
```

```python
df['T1-Runs'] = df['T1-Runs'].astype(int)
df['T2-Runs'] = df['T2-Runs'].astype(int)
df['T1-W'] = df['T1-W'].astype(int)
df['T2-W'] = df['T2-W'].astype(int)
```

```python
def win(x):
    if( x["T1-Runs"] > x["T2-Runs"]):
        return x["Team-1"]
    else:
        return x["Team-2"]
```

```python
df["Winner"] = df.apply(win,axis=1)
```

```python
df
```

```python
def Chase_defended(x):
    if( x["T1-Runs"] > x["T2-Runs"]):
        return "Defending"
    else:
        return "Chasing"
```

```python
df["Won-By"] = df.apply(Chase_defended,axis=1)
```

```python
df.drop("result",axis=1,inplace=True)
```

```python
df2 = df[['Team-1','Team-2','T1-Runs', 'T1-W', 'T2-Runs', 'T2-W','Winner','Man of the match','Won-By','venue', 'stadium','Date','Time']]
```
